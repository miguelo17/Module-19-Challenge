```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
#df_market_data.columns
numeric_columns = ['price_change_percentage_24h', 'price_change_percentage_7d',
       'price_change_percentage_14d', 'price_change_percentage_30d',
       'price_change_percentage_60d', 'price_change_percentage_200d',
       'price_change_percentage_1y']
scaler = StandardScaler()

scaled_data = scaler.fit_transform(df_market_data[numeric_columns])

scaled_df = pd.DataFrame(scaled_data, columns=numeric_columns)

scaled_df.head()
```

```python
# Create a DataFrame with the scaled data
scaled_df

# Copy the crypto names from the original DataFrame

crypto_names = df_market_data.index


# Set the coin_id column as index
scaled_df.set_index(crypto_names, inplace=True)

# Display the scaled DataFrame
df_market_data_scaled = scaled_df.copy()

df_market_data_scaled.head(10)
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 12))

k
```

```python
# Create an empty list to store the inertia values

inertia =[]

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for i in k:  
    model = KMeans(n_clusters=i, random_state=1)  # create a KMeans model with n_clusters set to the current value of i
    model.fit(df_market_data_scaled)  # fit the model to the scaled data
    inertia.append(model.inertia_)  # Append the inertia value to the inertia list

inertia
```

```python
# Create a dictionary with the data to plot the Elbow curve

elbow_data = {
    "k": k,          # k values
    "inertia": inertia  # corresponding inertia values
}

# Create a DataFrame with the data to plot the Elbow curve

df_elbow1 = pd.DataFrame(elbow_data)

df_elbow1
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.

df_elbow1.hvplot.line(
    x='k',                  # x-axis represents the values of k
    y='inertia',           # y-axis represents the inertia values
    title='Elbow Curve 1',   # Title of the plot
    xticks=k,
    line_color =  "orange"           
)

```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** In this grapgh we can see that the elbow angle forms much clearer in the k value of 4, i will select this to form the clusters based on this appreciation.

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k

best_k = 4  #depending on your analysis

# Initialize the K-Means model with the best value for k
model = KMeans(n_clusters=best_k, random_state=1)  # random_state is set for reproducibility

# Print the model to confirm initialization
model
```

```python
# Fit the K-Means model using the scaled DataFrame

model.fit(df_market_data_scaled)  # Fit the model using the scaled DataFrame

# Make predictions using the fitted model
stock_clusters = model.predict(df_market_data_scaled)

# Print the predicted clusters
stock_clusters
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame

crypto_clusters = model.predict(df_market_data_scaled)  # Predict the clusters for the scaled DataFrame


# Print the resulting array of cluster values.
print(crypto_clusters)
```

```python
# Create a copy of the scaled DataFrame
df_crypto_scaled_predictions = df_market_data_scaled.copy()  # Create a copy of the scaled DataFrame

# Print the new DataFrame to confirm the copy
df_crypto_scaled_predictions.head()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters

df_crypto_scaled_predictions['Cluster'] = crypto_clusters


# Display the copy of the scaled DataFrame

df_crypto_scaled_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.

scatter_plot1 = df_crypto_scaled_predictions.hvplot.scatter(
    x="price_change_percentage_24h",  # x-axis
    y="price_change_percentage_7d",    # y-axis
    by="Cluster",                       # Color by cluster labels
    hover_cols=[df_crypto_scaled_predictions.index.name]             # Show cryptocurrency names on hover
)

scatter_plot1
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.

pca = PCA(n_components=3)

pca
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.

pca_data = pca.fit_transform(df_market_data_scaled)

pca_df = pd.DataFrame(pca_data, columns=["PC1", "PC2", "PC3"])

# View the scaled PCA data

pca_df.head()
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_market_data_scaled)

pca = PCA(n_components=3)  
pca.fit(scaled_data)

explained_variance = pca.explained_variance_ratio_

print("Explained Variance Ratio:", explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** These three components together capture 89.50% of the total information in the dataset, since the first two components together explain around 72% of the variance, they are the most significant.

```python
# Create a new DataFrame with the PCA data.

pca_df

# Copy the crypto names from the original scaled DataFrame

crypto_names = df_market_data.index


# Set the coin_id column as index
pca_df.set_index(crypto_names, inplace=True)


# Display the scaled PCA DataFrame

df_market_pca = pca_df.copy()

df_market_pca.head()
```

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11

k_values = list(range(1, 12))

k
```

```python
# Create an empty list to store the inertia values

inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list

for i in k:  
    model = KMeans(n_clusters=i, random_state=1)  # create a KMeans model with n_clusters set to the current value of i
    model.fit(df_market_pca)  # fit the model to the scaled data
    inertia.append(model.inertia_)  # Append the inertia value to the inertia list

inertia
```

```python
# Create a dictionary with the data to plot the Elbow curve

elbow_data = {
    "k": k,
    "inertia": inertia
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow2 = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow2
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.

elbow_plot2 = df_elbow2.hvplot.line(
    x='k',
    y='inertia',
    title='Elbow Curve',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    xticks=k
)

elbow_plot2
```

* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** It is likely similar since the elbow presents at the same rate/values from the original data.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k

best_k = 3

# Initialize the K-Means model
kmeans_model = KMeans(n_clusters=best_k, random_state=1)

# Print the model to confirm initialization
kmeans_model
```

```python
# Fit the K-Means model using the PCA data

kmeans_model.fit(df_market_pca)

# Make predictions using the fitted model
stock_clusters = kmeans_model.predict(df_market_pca)


df_market_pca['StockCluster'] = stock_clusters

# Display the first few rows
df_market_pca.head()

```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.

scatter_plot2 = df_market_pca.hvplot.scatter(
    x='PC1',
    y='PC2',
    by='StockCluster',  # Color by the K-Means cluster labels
    hover_cols=[df_market_pca.index.name], 
    title='PCA Scatter Plot of Cryptocurrency Clusters',
    xlabel='Principal Component 1 (PC1)',
    ylabel='Principal Component 2 (PC2)'
  
)

scatter_plot2

```
